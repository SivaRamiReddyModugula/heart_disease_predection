```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
style.use('ggplot')
import seaborn as sns
plt.rcParams['figure.figsize']=(15,10)
```

```python
df=pd.read_csv("heart.csv")
df.head()
```

```python
df.hist()
```

```python
df.columns
```

```python
labels=['trestbps', 'chol', 'thalach', 'oldpeak']
```

```python
for label in labels:
    print(label)
    print(df[label].describe())
    print(df[label].unique())
    sns.boxplot(df[label])
    plt.show()
```

```python
# Handling Out_layers
# IQR

df['trestbps']=np.where(df['trestbps']>=140,140,df['trestbps'])
df['chol']=np.where(df['chol']>=275,275,df['chol'])
df['thalach']=np.where(df['thalach']<=132,132,df['thalach'])
df['oldpeak']=np.where(df['oldpeak']>=1.8,1.8,df['oldpeak'])
```
